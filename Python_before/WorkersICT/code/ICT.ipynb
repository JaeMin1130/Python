```python
raw = pd.read_excel(r'C:\Users\user\OneDrive\바탕 화면\study_22-ICT\3_workersICT\data\ICT_data.xlsx')

# survey features labeling(A_, B_, C_, D_, E_)
def survey(label, x, y):
    raw.iloc[1, range(x,y)] = label + raw.iloc[1, range(x,y)]
survey('A_', 92, 114)
survey('B_', 115, 167)
survey('C_', 168, 192)
survey('D_', 193, 240)
survey('E_', 241, 249)

# push columns, remove index[[0, 1]] -> 한 번에 할 수 있는 방법 없나
raw.columns = raw.iloc[1, :].array
raw.drop(raw.index[[0,1]], axis=0, inplace=True)
raw.index = range(0,346)

# remove columns which values are all Nan
dropped_columns=[]
for col in raw.columns:
    if raw[col].isna().all():
        dropped_columns.append(col)

print(dropped_columns)
first_df = raw.drop(dropped_columns, axis=1, inplace=False)
```

#### BIRTH_DATE -> AGE


```python
# 나이 계산기
from datetime import date
def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

first_df.BIRTH_DATE.isnull().any() 
first_df.rename(columns = {'BIRTH_DATE':'AGE'}, inplace=True)

# 자료형을 date형으로 바꾸고, 나이 계산해서 'AGE' 열에 넣기
age = []
for i in range (0, 346):
    a = '{:.0f}'.format(first_df.AGE[i])
    year = int(a[0:4])
    month = int(a[4:6])
    day = int(a[6:])

    if month > 12 : print(i)
    else: b = date(year, month, day)
    c = calculate_age(b)
    age.append(c)

first_df.AGE = age
```
