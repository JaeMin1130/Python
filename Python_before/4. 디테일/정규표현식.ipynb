```python
# 반복  *, +, {a}, {a,}(a 이상), {,a}(a 이하), {a,b}
```

```python
import re
p=re.compile('[a-z]+')
m= p.match('favorite')   # match는 처음부터 매치되는지 그래서 'No match'
if m:
  print('Match found', m.group(),m.start(), m.end(), m.span())
else: print('No match')
```

```python
m= re.search('[vorit]+','favorite')   # search는 전체에서 매치되는지 그래서 'avor', re.compile과 동시에 실행
if m:
  print('Match found', m.group(),m.start(), m.end(), m.span())
else: print('No match')
```

```python
p=re.compile('[a-z]+')
result=p.findall('life is tooshort')  # findall은 공백을 기준으로 문자열을 나누어 리스트로 리턴
print(result)  

result1=p.finditer('life istoo short') # finditer은 반복 가능한 객체를 리턴
print(result1)
for i in result1: print(i)
```

```python
# 컴파일 옵션
import re
p=re.compile('[a.e]',re.S)  # re.DOTALL 줄바꿈 포함 매치
m=p.match('abc')
print(m)
n=re.search('[a.b]','aBd', re.I) # re.IGNORECASE 대,소문자 구분 없이 매치
print(n)
```

```python
p=re.compile('^python+',re.M)   # re.MULTILINE 여러 줄과 매치(^, $ 와 같이 사용)
data='''python one life is too short
python two you need python 
python three'''

print(p.findall(data))
```

```python
p=re.compile('crowl|dog')    # a|b   a 또는 b 
print(p.search('hellocrowl'))

print(re.search('^life', 'life is too short'))   # ^a   문자열의 맨 처음이 'a'여야 매치

print(re.search('[short$]','life is too short'))   # a$   문자열의 맨 마지막이 'a'여야 매치
```

```python
p=re.compile(r'\bclass\b')    # r과 함께 사용, \b 단어 구분자, \B는 공백이 하나도 없음을 의미
print(p.search('no class at all'))
print(p.search('classify'))
```
