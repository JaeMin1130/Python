```python
dir([1,2])    # dir() 객체가 가지고 있는 변수나 함수를 보여준다.
for a,dog in enumerate(['body','foo','bar']):  # enumerate() 객체의 위치를 알고 싶을 때 사용 
  print(a,dog)
a=[lambda a,b:a+b, lambda a,b:a*b]   # lambda 간단한 함수 만들 때, 리스트 내에 만들 때 사용
len('puddle') # len() 입력값의 길이를 리턴
list(3*a for a in [1,2,3])  # list() 리스트로 만듦 * set(), tuple()

def dog(x): return x*2
list(map(dog,[1,2,3,4]))  # map(함수명,입력인수) 입력인수를 함수에 대입한 값을 알려줌
list(map(lambda a:a*2,[1,2,3,4]))

#open(dog.py,'r') # open(파일명, 모드) *모드: 'r', 'w', 'a', 'b'
list (range(0,10,3)) # range(a,b,c)  세 번째 인수(c)는 거리
a=[1,2,3]
b=[4,4,5,6,7]
c=1,2
list(zip(a,b,c))  # zip(a,b,c) a,b,c 각각 동일 위치에 있는 요소끼리 묶어줌
```
