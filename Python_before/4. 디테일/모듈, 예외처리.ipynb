```python
# try: except: 수행할 문장
try: 
  a=1,2,3 
  a[4]
except: print(sum([1,2,3]))
```

```python
# try: except 발생 오류: 수행할 문장
try: 
  a=1,2,3 
  a[4]
except IndexError: pass     # 오류가 발생하면 그냥 통과
```

```python
# try: except 발생 오류 as 오류 메세지 변수:
try: 
  a=1,2,3 
  a[4]
except IndexError as c: print(sum([2,3,4]))
```

```python
# try...else
try: 'a'+'1'
except TypeError as d: print('You are not my type.')
else: print('a'*2)

# try...finally
```

```python
# 오류 일부러 발생시키기
class Service:
  def __init__(self,name):
    self.name=name
  def sum(self,a,b):
    result=a+b
    print('%s님 %s+%s = %s입니다.'%(self.name , a,b,result))
    raise NameError          # 오류를 발생시키고 싶은 함수 안에 'raise 오류명'
class dog(Service): pass

coco = dog('puddle')
coco.sum(3,4)
```
