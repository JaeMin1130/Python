```python
# If문 Q1
a='Life is too short, you need python'
if'wife' in a:
  print('wife')
elif'python' in a and 'you' not in a:
  print('python')
elif 'shirt' not in a:
  print('shirt')
elif'need' in a:
  print('need')
else:
  print('none')

# while문 Q1
b=0
while b<5:
  b=b+1
  print('*'*b)

while True:        # 대문자 'T'
  b+=1             # while문 수행 시 1씩 증가
  if b>5:break     # 5보다 커지면 벗어남
  print('*'*b)

# for문 Q1
a=[70,60,55,75,95,90,80,80,85,100]
total=0
for score in a:
  total+=score         # total=total+score
average= total/len(a)  
print(average)
```
