```python
# for 변수 in list(tuple, string)
# 5명의 학생이 시험을 보았는데 시험 점수가 60점이 넘으면 합격이고 그렇지 않으면 불합격이다. 합격인지 불합격인지 결과를 보여주시오.
number=0
for test_score in [67,58,10,77,90]:
  number=number+1
  if test_score > 60: print('%d student is pass, %d points over.' %(number, (test_score-60)))
  else: print('%d student is fail, %d points lower.' %(number,(60-test_score)))
```

```python
# continue
number=0
a= [67,58,10,77,90]
for test_score in a:
  number=number+1
  if test_score<60: continue
  else:print('no.%d student congratulation!'%number )

for number in range(len(a)):  # len(): 리스트 내 요소의 갯수(5)
  if a[number] < 60: continue
  else:print('no.%d student congratulation!'%(number+1) )
```

```python
# range 함수
range(0,10)  # 0~9, 'range()' 자체가 객체
sum=0
for i in range(0,11):
  sum=i+sum
  print(sum)      # for문 과정을 다 보여줌
print(sum)        # for문 최종결과를 보여줌

#구구단
for i in range(2,10):
  for j in range(1,10):
    print(i*j, end=' ')  # 입력인수 'end': 한 줄에 계속 출력하려고
  print('')              # 두 번째 for문이 끝나면 다음 줄부터 출력하게 해줌
 
```

```python
# 리스트 내포
a=[1,2,3,4]
b=[]
for i in a:
  b.append(i*3)     # .append():리스트 요소 추가
print(b)

b=[i*3 for i in a]  # 리스트 내포
print(b)

b=[i*3 for i in a if i % 2==0]  # if문 추가
print(b)

c=[i*j for i in range(2,10)
          for j in range (1,10)]
print(c)
```
