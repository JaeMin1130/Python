```python
# 조건문이 참인 동안에 while문 아래에 속하는 문장들이 반복해서 수행된다
# 열 번 찍어 안 넘어가는 나무 없다
treehit=0
while treehit<10:
  treehit=treehit+1
  print('나무를 %d번 찍었습니다.' %treehit)
if treehit==10:
  print('나무 넘어갑니다.')
```

```python
prompt='''
1. Cat
2. Hedgehog
3. Fish
4. Dog
'''
number=0
while number!=4:
  print(prompt)
  number=int(input('Enter number:'))   # 4 칠 때까지 무한반복
```

```python
# coffee machine(break 사용)
coffee=5
while True:
 money=int(input('insert coin: '))
 if money == 300:
   print('coffee')
   coffee=coffee-1
 elif money>300: 
   print('change %d and coffee' %(money-300))
   coffee=coffee-1
 else:
   print('not enough money')
   print('%d coffee left' %coffee)
 if not coffee:
   print('sold out')
   break
```

```python
#continue(처음 조건문으로 돌아가게 할 때)
a=0
while a<10:
  a=a+1
  if a%2 == 0:continue  
  print(a)
```
