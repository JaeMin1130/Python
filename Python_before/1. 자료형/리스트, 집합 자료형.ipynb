```python
#리스트 자료형
a=[]    #a=list()도 가능
b=[1, 2, 'I\'m', 'hungry']
c=[1, 2, ['I\'m', 'hungry']]
```

```python
#리스트의 인덱싱
a=[1, 2, 3]
a[0]+a[2]    # = 4
b=['Life','is','too','short']
c=b[0]*a[1]    #'LifeLife'
c.replace('eL','e L')  #'Life Life'
d=[1, 2, 'I have a ', ' dog', ['pretty', 'cute']]
d[2]+d[4][0]+d[-2]   #'I have a pretty dog' 
```

```python
#리스트의 슬라이싱
a=[1, 2, 3, [4, 5]]
a[0:3]    #[1, 2, 3]
a[3][:1]  #[4]        
```

```python
#리스트의 연산자
a=[1,2,3]
b=[6,7,8]
a+b     #[1,2,3,6,7,8]
a[0:2]+b[:2]  #[1,2,6,7]

b*a[1]  #[6,7,8,6,7,8]
```

```python
#리스트 수정
a=[1,2,3]
a[2]=4    #a=[1,2,4]
a[0:2]=['a','b']  #a=['a','b',4]
a[0]=['c','d']    #a=[['c','d'],'b',4]

#리스트 삭제
a[:1]=[]   #a=['b',4]
del a[1]   #a=['b']
a=[1,2,3,4]
del a[2:]  #a=[1,2]
```

```python
#리스트 관련 함수들
a=[2,4,3,1]
a.append(4)     #리스트 요소 추가 a=[2,4,3,1,4]
a.sort()        #리스트 정렬   a=[1,2,3,4,4]
a.reverse()     #리스트 뒤집기 a=[4,4,3,2,1]
a.index(3)      #위치 반환  = 2
a.pop()         #요소 끄집어내기(위치값) = 1, a=[4,4,3,2]
a.count(4)      #요소 x의 갯수   = 2
a.extend([4,5]) #리스트 확장 a=[4,4,3,2,4,5]
```

```python
# list(), tuple(), set() 간 변환 가능
a=[1,2,3,4,3,2,1]
set(a)       #리스트에서 중복된 요소 지울 수 있다.
```

```python
#집합 자료형
s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9}
#교집합 {4,5,6}
s1&s2  
s1.intersection(s2)
#합집합 {1,2,3,4,5,6,7,8,9}
s1|s2
s1.union(s2)
#차집합 {1,2,3}
s1-s2
s1.difference(s2)
```

```python
#집합 자료형 관련 함수들
s1=set([1,2,3])
s1.add('dog')      #값 1개 추가하기 {1, 2, 3, 'dog'}
s1.update({4,5,6}) #값 여러 개 추가하기 {1, 2, 3, 4, 5, 6, 'dog'}
s1.remove('dog')   #값 1개 제거하기 {1, 2, 3, 4, 5, 6}
```
