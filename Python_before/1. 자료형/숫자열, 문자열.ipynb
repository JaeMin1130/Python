```python
#문자열 인덱싱(파이썬은 0부터 숫자를 센다.)               
a='Life is too short, You need Python.'
a[0]        #'L'
a[6]        #'s'
print(a[6]) # s
a[-3]       #'o'

#문자열 슬라이싱(끝번호 미포함)
a[0:4]           #'Life'
a[8:]            #'too short, You need Python.'
a[:8]            #'Life is '
a[19:-7]         #'You need '

a='20211222Sunny'
Year=a[:4]      #'2021'
Date=a[4:8]     #'1221'
Weather=a[8:]   #'Sunny'

#'Pithon'을 'Python'으로 바꾸기
a='Pithon'
a=a[:1]+'y'+a[2:]
print(a)
```

```python
#문자열 포매팅
#I have 3(three) dogs
a=3
b='three'
'I have %d dogs' %3        # %a
'I have %s dogs' %3        # %'three' %a %b 
'I have %s%% dog' %87      # '%'를 쓰고 싶을 때
'I have %0.0f dogs' %3     # 소수점 0번 째 자리까지만 표현하고 싶을 때
'I have %-9d dogs' %3124   # 왼쪽 정렬, 문자도 가능
'I have %10s dogs' %51251  # 오른쪽 정렬, 문자도 가능
'I have %5.2f dogs'%3.2345 # 오른쪽 정렬+소수점 2번 째 자리까지 표현
'I\'m %d years old, %s next year' %(25,26)  #여러 개 넣을 때
```

```python
#고급 문자열 포매팅
'I have {0} dogs, {1} cats' .format(4,5)
'I have {0} dogs, {cute} cats' .format(6, cute=5)
'{0:<10}{1:^10}{2:>10}'.format('hi','bye','miss') #왼쪽 정렬, 가운데 정렬, 오른쪽 정렬
'{0:ㅠ^10}'.format('ㅛ')     #가운데 정렬, 공백 채우기
'{0:10.2f}'.format(3.141592) #소수점 표현, 자릿수 맞추기
```

```python
#문자열 관련 함수들
a='Life is too short'
a.count('i')  #.count() 문자 개수 세기
a.find('t')   #.find() or.index() 찾기
a.join('dog')  #.join 요소들 사이사이에 변수 a 삽입 
               # = dLife is too shortoLife is too shortg'
a.upper()     #.upper 대문자로 변환
a.lower()     #.lower 소문자로 변환
a.lstrip()    #왼쪽 공백 지우기
a.rstrip()    #오른쪽 공백 지우기
a.strip()     #양쪽 공백 지우기
a.replace('short','long')  #문자열 바꾸기
a.split()     #괄호 안의 문자를 기준으로 문자열 나누기
```
