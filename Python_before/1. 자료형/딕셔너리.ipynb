```python
#딕셔너리(key와 value) 
a={1:'x'}
a[2]='life'   #쌍 추가
a['x']=[1,2]  #{1: 'x', 2: 'life', 'x': [1, 2]}

del a[1]      #삭제 key가 1인 쌍 {2: 'life', 'x': [1, 2]}

a['x']        #key가 'x'인 value 반환 [1,2]

# *key 중복 금지  key에 튜플은 되고 리스트는 안됨
```

```python
#딕셔너리 관련 함수들
a={'name':'JaeMin','phone':'01051672315','birth':'1130'}
a.keys()  #key 리스트 만들기    dict_keys(['name', 'phone', 'birth'])
for b in a.keys():
  print(b)
list(a.keys())       #리스트로 변환 ['name','phone','birth']
a.get('name')        #a['name']과 동일, key로 value얻기
a.get('dog','cute')  #'dog'가 없을 때 'cute'리턴
'weather' in a       #key가 있는지 조사하기 True or False
'name' in a
```
