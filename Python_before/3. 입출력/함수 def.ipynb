```python
# 함수(def...return)
def f(a,b):
  return a+b

# 입력인수가 불특정 다수일 때 튜플 형태로 들어온다.
def lot(*args):           # 입력변수 앞에 '*'을 붙인다
  sum=0
  for i in args:
    sum+=i
  return sum
c=lot(2,4,6,8)
print(c)

# 입력인수가 (string,integer)일 때
def dog(letter,*args):
  if letter=='pretty':
    sum=0
    for i in args:
      sum += i
  elif letter == 'cute':
    sum=1
    for i in args:
      sum *= i
  return sum
d=dog('cute',2,3,4)
print(d)
```
