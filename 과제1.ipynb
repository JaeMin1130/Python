```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

# y = f(x)
def f(x):
    return x**4 - 3*x**3 +2

# 미분 함수
def fprime(x):
    h = 0.001
    return (f(x+h) - f(x))/h

# y = f(x) 정의역, 치역
all_x = tf.linspace(-5, 5, 100)
all_y = f(all_x)

x = tf.Variable(0.1) # initial value
LR = 0.1
x_history, y_history = [], []

for _ in range(100):
    y = f(x)
    step = LR * fprime(x)

    x_history.append(x.numpy())
    y_history.append(y.numpy())

    x.assign_sub(step)  # x = x - lr*dx

fig, ax = plt.subplots(figsize=(5, 5))

ax.plot(all_x, all_y) # y = f(x) 그리기
ax.scatter(x_history, y_history) # 미분 history

plt.xlabel('$x$')
plt.ylabel('$y=x^2-2x+1$')
plt.show()
```
