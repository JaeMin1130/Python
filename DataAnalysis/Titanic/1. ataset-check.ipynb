```python
df_train = pd.read_csv('./Titanic_data/train.csv')
df_test = pd.read_csv('./Titanic_data/test.csv')
df_gender= pd.read_csv('./Titanic_data/gender_submission.csv')
```

```python
df_train.describe()    # .describe() 간단한 통계값 알려줌
```

```python
df_test.describe()
```

```python
# 각 columns에 null값 파악하기
for col in df_train.columns:
    msg = 'column: {:>10}\t Percent of NaN value: {:.2f}%'.format(col, 100*(df_train[col].isnull().sum() / df_train[col].shape[0]))
    print(msg)
```

```python
# null values matrix로 나타내기

msno.matrix(df=df_train.iloc[:,:], figsize=(8,8), color=(0.8, 0.5, 0.2))  # .matrix 대신 .bar 가능
msno.matrix(df=df_test.iloc[:,:], figsize=(8,8), color=(0.8, 0.5, 0.2))
```

```python
# df_train 'Survived' 열 데이터 그래프로 나타내기
f, ax = plt.subplots(1, 2, figsize=(20,8))   # 도화지
                                            # explod = 짼다    # autopercentage   # 첫 번째 열
df_train['Survived'].value_counts().plot.pie(explode=[0,0.1], autopct='%1.1f%%', ax=ax[0], shadow=True)
ax[0].set_title('Pie plot - Survived')
ax[0].set_ylabel('')

sns.countplot('Survived', data=df_train, ax=ax[1])
ax[1].set_title('Count plot - Survived')
ax[1].set_ylabel('')
plt.show()
```
